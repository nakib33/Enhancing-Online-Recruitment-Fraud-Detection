```python
import re
import operator
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
```

```python
def remove_stopwords(tweet):
    stop_words = set(stopwords.words('english'))
    words = tweet.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
def clean_text(text):
    text = text.lower()
    text = remove_stopwords(text)
    # Remove numeric data
    text = re.sub(r"\d+", "", text)
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    # Remove special characters and punctuation
    text = re.sub(r"[^\w\s]", "", text)
    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text)
    # Remove Non-ASCII
    text = ''.join(char for char in text if ord(char) < 128)
    return text.strip()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler

from transformers import AutoTokenizer, TFAutoModel
```

```python

new_df["text_c"] = new_df["text_c"].astype(str)

X_text = new_df["text_c"].values
y = new_df["fraudulent"].values.astype(int)

X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
ros = RandomOverSampler(random_state=42)


X_train_text_res, y_train_res = ros.fit_resample(
    X_train_text.reshape(-1, 1),
    y_train
)

# back to 1D text array
X_train_text_res = X_train_text_res.ravel()


```

```python
MODEL_NAME = "bert-base-uncased"
MAX_LEN = 128
BATCH_SIZE = 12

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

train_enc = tokenizer(
    X_train_text_res.tolist(),
    truncation=True,
    padding="max_length",
    max_length=MAX_LEN,
    return_tensors="tf"
)

test_enc = tokenizer(
    X_test_text.tolist(),
    truncation=True,
    padding="max_length",
    max_length=MAX_LEN,
    return_tensors="tf"
)

y_train_tf = tf.convert_to_tensor(y_train_res, dtype=tf.int32)
y_test_tf  = tf.convert_to_tensor(y_test, dtype=tf.int32)
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((
    {"input_ids": train_enc["input_ids"], "attention_mask": train_enc["attention_mask"]},
    y_train_tf
)).shuffle(len(y_train_res), seed=42).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((
    {"input_ids": test_enc["input_ids"], "attention_mask": test_enc["attention_mask"]},
    y_test_tf
)).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print("Train samples:", len(y_train_res))
print("Test samples:", len(y_test))
```

```python
import tensorflow as tf

VOCAB_SIZE = 30522
EMBED_DIM  = 128

class MeanPooling(tf.keras.layers.Layer):
    def call(self, inputs):
        x, attention_mask = inputs  # x: (B,T,E), mask: (B,T)
        mask = tf.cast(tf.expand_dims(attention_mask, axis=-1), tf.float32)  # (B,T,1)
        x = x * mask
        summed = tf.reduce_sum(x, axis=1)                 # (B,E)
        counts = tf.reduce_sum(mask, axis=1)              # (B,1)
        counts = tf.maximum(counts, 1e-9)
        return summed / counts

# Inputs
input_ids = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32, name="input_ids")
att_mask  = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32, name="attention_mask")

# Embedding
x = tf.keras.layers.Embedding(
    input_dim=VOCAB_SIZE,
    output_dim=EMBED_DIM,
    mask_zero=False
)(input_ids)


emb = MeanPooling()([x, att_mask])

# Classifier
h = tf.keras.layers.Dense(16, activation="relu")(emb)
h = tf.keras.layers.Dense(32, activation="relu")(h)
logits = tf.keras.layers.Dense(2)(h)

model = tf.keras.Model(
    inputs={"input_ids": input_ids, "attention_mask": att_mask},
    outputs=logits
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-6),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=50
)

test_loss, test_acc = model.evaluate(test_ds)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_training_curves(history, acc_keys=("accuracy", "acc")):
    h = history.history


    train_acc_key = next((k for k in acc_keys if k in h), None)
    val_acc_key   = next((f"val_{k}" for k in acc_keys if f"val_{k}" in h), None)


    plt.figure(figsize=(7,4))
    plt.plot(h.get("loss", []), label="train_loss")
    if "val_loss" in h:
        plt.plot(h["val_loss"], label="val_loss")
    plt.title("Training vs Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.legend()
    plt.show()


    if train_acc_key is not None:
        plt.figure(figsize=(7,4))
        plt.plot(h[train_acc_key], label="train_accuracy")
        if val_acc_key is not None:
            plt.plot(h[val_acc_key], label="val_accuracy")
        plt.title("Training vs Validation Accuracy")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.grid(True)
        plt.legend()
        plt.show()
    else:
        print("No accuracy key found in history. Available keys:", list(h.keys()))


plot_training_curves(history)
```

```python
# Inputs
input_ids = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32, name="input_ids")
att_mask  = tf.keras.Input(shape=(MAX_LEN,), dtype=tf.int32, name="attention_mask")

# Embedding Layer (RoBERTa vocab)
x = tf.keras.layers.Embedding(
    input_dim=VOCAB_SIZE,
    output_dim=EMBED_DIM,
    mask_zero=False
)(input_ids)

# Mean Pooling
emb = MeanPooling()([x, att_mask])

# Classifier
h = tf.keras.layers.Dense(16, activation="relu")(emb)
h = tf.keras.layers.Dense(32, activation="relu")(h)
logits = tf.keras.layers.Dense(2)(h)

model = tf.keras.Model(
    inputs={"input_ids": input_ids, "attention_mask": att_mask},
    outputs=logits
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-6),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=50
)

test_loss, test_acc = model.evaluate(test_ds)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import matplotlib.pyplot as plt

def plot_training_curves(history, acc_keys=("accuracy", "acc")):
    h = history.history


    train_acc_key = next((k for k in acc_keys if k in h), None)
    val_acc_key   = next((f"val_{k}" for k in acc_keys if f"val_{k}" in h), None)


    plt.figure(figsize=(7,4))
    plt.plot(h.get("loss", []), label="train_loss")
    if "val_loss" in h:
        plt.plot(h["val_loss"], label="val_loss")
    plt.title("Training vs Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.legend()
    plt.show()


    if train_acc_key is not None:
        plt.figure(figsize=(7,4))
        plt.plot(h[train_acc_key], label="train_accuracy")
        if val_acc_key is not None:
            plt.plot(h[val_acc_key], label="val_accuracy")
        plt.title("Training vs Validation Accuracy")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.grid(True)
        plt.legend()
        plt.show()
    else:
        print("No accuracy key found in history. Available keys:", list(h.keys()))

plot_training_curves(history)
```

```python
import numpy as np
import tensorflow as tf

# Get logits
y_pred_logits = model.predict(test_ds)

# Convert logits → class labels (0 or 1)
y_pred = np.argmax(y_pred_logits, axis=1)

# True labels
y_true = np.concatenate([y for x, y in test_ds], axis=0)

print("Pred shape:", y_pred.shape)
print("True shape:", y_true.shape)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt

y_true_all = []
y_pred_all = []

for batch_x, batch_y in test_ds:
    logits = model(batch_x, training=False)
    preds  = tf.argmax(logits, axis=1).numpy()
    y_pred_all.append(preds)
    y_true_all.append(batch_y.numpy())

y_true = np.concatenate(y_true_all)
y_pred = np.concatenate(y_pred_all)

print("Test Accuracy:", accuracy_score(y_true, y_pred))
print(classification_report(y_true, y_pred, target_names=["0","1"]))
```
